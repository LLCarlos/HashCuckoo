# Predicition mechanisms - Tarde-offs - False Positives vs Reaction Time

### Figure 8(d)

```python
#-*- coding: UTF-8 -*-
"""
========
Predicition mechanisms - Tarde-offs - False Positives vs Reaction Time
========

"""
import numpy as np
import matplotlib.pyplot as plt
import pylab as pyl
from matplotlib import rc

rc('text', usetex=True)
plt.rcParams["font.family"] = "Times New Roman"

ticks = [5+10*i for i in range(9)]

SIZE=16

fig, ax = plt.subplots(figsize=(7, 4))

fp_cp = [92.3, 90.1, 84.5, 75.7, 55.2, 31.7, 15.0,  6.0, 5.0]
fp_dp = [94.9, 94.2, 92.7, 89.7, 81.3, 63.8, 45.0, 32.3, 20.1]

ax.plot(ticks, fp_cp, color='#388E3C', marker='s', linestyle='--', label='FP - Control Plane')
ax.plot(ticks, fp_dp, color='#66bb6a', marker='^', linestyle='--', label='FP - Data Plane')

plt.xticks(fontsize=SIZE)
plt.yticks(fontsize=SIZE)
ax.set_ylabel('False Positives - FP (\%)' , fontsize=SIZE,  color = 'black')
ax.set_xlabel('Number of Samples (previous flows)' , fontsize=SIZE)

ax.set_xticks(ticks)
xlabels = [2**i for i in range(4,13)]
ax.set_xticklabels((xlabels), fontsize=SIZE-4, rotation=0)

ax.set_yticks(np.arange(0, 121,20))

ax.set_axisbelow(True)
ax.yaxis.grid(color='silver', linestyle='-', linewidth=0.4, alpha=0.7)
ax.xaxis.grid(color='silver', linestyle='-', linewidth=0.4, alpha=0.7)

ax.tick_params(direction='inout', length=10)
ax.tick_params(axis='y', labelcolor='black')

#################################################################################################

ax2 = ax.twinx() 
ax2.set_yticks(np.arange(0, 122,20))

rt_cp = [32.8, 32.9, 34.2, 34.5, 37.8, 42.2, 68.1, 90.3, 102.3]
rt_dp = [17.8, 17.9, 18.1, 17.9, 19.2, 20.2, 28.6, 38.0,  43.1]

ax2.plot(ticks, rt_cp, color='#007c91', marker='o', linestyle='--', label='RT - Control Plane')
ax2.plot(ticks, rt_dp, color='#4dd0e1', marker='x', linestyle='--', label='RT - Data Plane')

ax.set_ylim(ymin=-5, ymax=130)
ax2.set_ylim(ymin=-5, ymax=130)

ax2.set_ylabel('Reaction Time - RT (ms)' , fontsize=SIZE, color = 'black')
plt.yticks(fontsize=SIZE)
ax2.tick_params(direction='inout', length=10, labelcolor = 'black')
ax2.set_axisbelow(True)


ax.legend(fontsize=SIZE-4, loc=2, frameon=True, edgecolor='black' )
ax2.legend(fontsize=SIZE-4, loc=1, frameon=True, edgecolor='black')


fig.tight_layout()
plt.show()
fig.savefig("fig/trade-offs.pdf")
```
