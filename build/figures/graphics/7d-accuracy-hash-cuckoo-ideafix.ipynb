# Hash Analysis - HashCuckoo and IDEAFIX Accuracy

#### Figure 7(d)

```python
#-*- coding: UTF-8 -*-
"""
========
Hash Analysis - HashCuckoo and IDAFIX Accuracy
========

"""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
import pylab as pyl
from matplotlib import rc
import pickle as p

rc('text', usetex=True)
plt.rcParams["font.family"] = "Times New Roman"

N = 5
SIZE=18

##### HashCuckoo ##################
FP_s2 = [ 53.971, 41.452, 35.031, 28.012, 26.201]   #S2 - False Positives
FN_s2 = [ 33.240, 21.180, 18.201, 16.123, 15.510]   #S2 - False Negatives
FP_s1 = [ 38.480, 26.530, 22.600, 17.770, 17.110]   #S1 - False Positives
FN_s1 = [ 18.197, 10.459,  8.243,  7.021,  6.902]   #S1 - False Negatives
 
##### IDEAFIX ################
FP_ideafix = [ 22.480, 14.530,  8.600,  3.770,  1.810]   #IDEAFIX - False Positives
FN_ideafix = [  10.197,  3.459,  1.243,  1.102,  0.095]   #IDEAFIX - False Negatives


ind = np.arange(N)  # the x locations for the groups
width = 0.6    # the width of the bars

fig, ax = plt.subplots(figsize=(7, 4))


##### False Positives ####
plt.plot(ind, FP_s2, color='r',       marker='^', linestyle='--', label='False Positives - HashCuckoo S2')
plt.plot(ind, FP_s1, color='#388E3C', marker='s', linestyle='--', label='False Positives - HashCuckoo S1')
plt.plot(ind, FP_ideafix, color='#007c91', marker='D', linestyle='--', label='False Positives - IDEAFIX')


##### False Negatives ####
plt.plot(ind, FN_s2, color='#ff9800', marker='o', linestyle='--', label='False Negatives - HashCuckoo S2')
plt.plot(ind, FN_s1, color='#66bb6a', marker='x', linestyle='--', label='False Negatives - HashCuckoo S1')
plt.plot(ind, FN_ideafix, color='#4dd0e1', marker='P', linestyle='--', label='False Negatives - IDEAFIX')


#S1 - False Positives text
ax.text(0.25, (FP_ideafix[0]),''.join(str(int(FP_ideafix[0]))+"\%"),
                ha='center', va='bottom', color='#007c91', fontsize=SIZE-6)


#S1 - False Negatives
ax.text(0.15, (FN_ideafix[0]+1),''.join(str(int(FN_ideafix[0]))+"\%"),
                ha='center', va='bottom', color='#4dd0e1', fontsize=SIZE-6)


# add some text for labels, title and axes ticks
plt.xticks(fontsize=SIZE)
plt.yticks(fontsize=SIZE)

ax.set_ylabel('Flows Identified (\%)' , fontsize=SIZE)
ax.set_xlabel('Memory Space for Hash Mapping (\%)' , fontsize=SIZE)
ax.set_xticks([0,1,2,3,4,])
ax.set_xticklabels(('50', '75', '100', '125', '150'), fontsize=SIZE, rotation=0)
ax.set_yticks(np.arange(0, 101,20))
ax.legend(fontsize=SIZE-7, ncol=2,loc=1, edgecolor='black')


ax.set_axisbelow(True)
ax.yaxis.grid(color='silver', linestyle='-', linewidth=0.4, alpha=0.7)
ax.xaxis.grid(color='silver', linestyle='-', linewidth=0.4, alpha=0.7)

ax.tick_params(direction='inout', length=10)
ax.tick_params( which='minor', direction='out', length=5)

fig.tight_layout()

plt.show()
fig.savefig("fig/accuracy-hashCuckoo-Ideafix.pdf")
```
