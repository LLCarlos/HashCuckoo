# Predicition mechanisms - Valid Predictions

```python
"""
========
Predicition mechanisms - Valid Predictions - True Positives
========

"""
import numpy as np
import matplotlib.pyplot as plt
import pylab as pyl
from matplotlib import rc

rc('text', usetex=True)
plt.rcParams["font.family"] = "Times New Roman"

N = 6
SIZE=16

ticks = [5+10*i for i in range(10)]

#Scenario 1
S1_CP   = (78, 90,  95.8, 97.7, 98.4, 98.6, 98.7, 98.8, 98.9, 98.9)   #S1 - Control Plane
S1_DP     = (56, 70, 80.25, 86.5, 90.2,     94,   96,   97,   98, 98) #S1 - Data Plane


#Scenario 2
S2_CP   = ( 20, 22, 26, 34.26, 44.8, 62.3, 78, 88, 94, 96)           #S2 - Control Plane
S2_DP     = ( 5.1,   5.8,  7.26, 10.26, 18.3, 36.3, 60, 78, 89, 94)  #S2 - Data Plane

ind = np.arange(N)  # the x locations for the groups
width = 0.6         # the width of the bars
fig, ax = plt.subplots(figsize=(7, 4))


### C1
plt.plot(ticks, S1_CP, color='#4A91C7', marker='s', linestyle='--', label='S1 - Control Plane')
plt.plot(ticks,   S1_DP, color='#455A64', marker='x', linestyle='--', label='S1 - Data Plane')

calc1 = ((S1_CP[0]-S1_DP[0])*100)/S1_CP[0]
rects4 = ax.errorbar((3.5,),(((S1_CP[0]-S1_DP[0])/2)+S1_DP[0],), yerr=((S1_CP[0]-S1_DP[0])/2), ecolor='red', linestyle='-.',  mew=1,capsize=4, label=None)
ax.text(7, ((S1_CP[0]-S1_DP[0])/2)+S1_DP[0]-2,''.join(str(int(calc1))+"\%"),
                ha='center', va='bottom', color='red', fontsize=SIZE-4)

## C2
plt.plot(ticks, S2_CP, color='#5FBAFF', marker='o', linestyle='--', label='S2 - Control Plane')
plt.plot(ticks, S2_DP, color='#607D8B', marker='^', linestyle='--', label='S2 - Data Plane')

calc2 = ((S2_CP[0]-S2_DP[0])*100.0)/S2_CP[0]
rects5 = ax.errorbar((3.5,),(((S2_CP[0]-S2_DP[0])/2)+S2_DP[0],), yerr=((S2_CP[0]-S2_DP[0])/2), ecolor='red', linestyle='-.',  mew=1,capsize=4, label=None)
ax.text(7, ((S2_CP[0]-S2_DP[0])/2)+S2_DP[0]-4,''.join(str(int(calc2))+"\%"),
                ha='center', va='bottom', color='red', fontsize=SIZE-4)


# add some text for labels, title and axes ticks
plt.xticks(fontsize=SIZE)
plt.yticks(fontsize=SIZE)
ax.set_ylabel('True  positives (\%)' , fontsize=SIZE)
ax.set_xlabel('Prediction Interval Tolerance (\%)' , fontsize=SIZE)

ax.set_xticks(ticks)
ax.set_xticklabels(ticks, fontsize=SIZE, rotation=0)


ax.set_yticks(np.arange(0, 101,20))
ax.legend(fontsize=SIZE-4, edgecolor='black')

ax.set_axisbelow(True)
ax.yaxis.grid(color='silver', linestyle='-', linewidth=0.4, alpha=0.7)
ax.xaxis.grid(color='silver', linestyle='-', linewidth=0.4, alpha=0.7)

ax.tick_params(direction='inout', length=10)
ax.tick_params( which='minor', direction='out', length=5)


fig.tight_layout()
plt.show()
fig.savefig("fig/accuracyPrediction.pdf")
```
